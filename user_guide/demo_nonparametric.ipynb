# Non parametric estimators

```python
import numpy as np
import matplotlib.pyplot as plt
from relife.parametric_model import Gompertz
from relife.non_parametric_model import KaplanMeier, NelsonAalen
from relife.datasets import load_circuit_breaker
```

## Basics

```python
time, event, entry = load_circuit_breaker()

km = KaplanMeier()
km.fit(time, event, entry)
```

## Plots

```python
import matplotlib.pyplot as plt
km.plot.sf()
plt.show()
```

```python
km = KaplanMeier().fit(time, event, entry)
gompertz = Gompertz().fit(time, event, entry)
```

```python
na = NelsonAalen().fit(time, event, entry)
```

```python
na.plot.chf()
plt.show()
```
