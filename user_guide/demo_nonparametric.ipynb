# Non parametric estimators

```python
import numpy as np
import matplotlib.pyplot as plt
from relife.lifetime_model import Gompertz, KaplanMeier, NelsonAalen
from relife.data import load_circuit_breaker
```

## Basics

```python
time, event, entry = load_circuit_breaker()

km = KaplanMeier()
km.fit(time, event, entry)
```

## Plots

```python
import matplotlib.pyplot as plt

timeline, sf, se = km.sf(se=True)
plt.plot(timeline, sf, drawstyle="steps-post")
plt.show()
```

```python
km = KaplanMeier().fit(time, event, entry)
gompertz = Gompertz().fit(time, event, entry)
```

```python
na = NelsonAalen().fit(time, event, entry)
```
